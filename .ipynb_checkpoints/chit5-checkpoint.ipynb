```python
df = pd.read_csv('Mall_Customers.csv')
print('Dataset Loaded Successfully')
print(df.head())
print('Shape:', df.shape)
```

```python
df.columns = [c.strip().replace(' ', '_').replace('(', '').replace(')', '').replace('$','') for c in df.columns]
print('Columns after cleaning:', df.columns.tolist())
```

```python
print(df.columns.tolist())
```

```python
# Label Encoding
le = LabelEncoder()
if 'Genre' in df.columns:
    df['Gender_enc'] = le.fit_transform(df['Genre'])
    print("Label Encoding Done!")

# Select features for clustering (correct column names)
X = df[['Annual_Income_k', 'Spending_Score_1-100']]

# Data Transformation (Scaling)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Show sample data
df.head()
```

```python
possible_income_cols = [c for c in df.columns if 'Income' in c or 'annual' in c.lower()]
possible_spend_cols = [c for c in df.columns if 'Spending' in c or 'spend' in c.lower()]


income_col = possible_income_cols[0] if possible_income_cols else None
spend_col = possible_spend_cols[0]


X1 = df[[spend_col]].copy()
X2 = df[[income_col, spend_col]].copy() if income_col else df[[spend_col]].copy()


scaler1 = StandardScaler()
X1_scaled = scaler1.fit_transform(X1)
scaler2 = StandardScaler()
X2_scaled = scaler2.fit_transform(X2)
print('Data Scaled Successfully!')
```

```python
X1_train, X1_test = train_test_split(X1_scaled, test_size=0.3, random_state=42)
X2_train, X2_test = train_test_split(X2_scaled, test_size=0.3, random_state=42)
print('Train-Test Split Done!')
```

```python

def find_best_k(X):
    best_k, best_score = 0, -1
    for k in range(2, 11):  # test k from 2 to 10
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(X)
        score = silhouette_score(X, labels)
        print(f"k = {k}, Silhouette Score = {score:.4f}")
        if score > best_score:
            best_k, best_score = k, score
    print(f"\nBest k: {best_k} with Silhouette Score = {best_score:.4f}")
    return best_k
```

```python
def find_best_k(X):
    best_k, best_score = 0, -1
    for k in range(2, 11):  # test k values from 2 to 10
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
        labels = kmeans.fit_predict(X)
        score = silhouette_score(X, labels)
        print(f"k = {k}, Silhouette Score = {score:.4f}")
        if score > best_score:
            best_k, best_score = k, score
    print(f"\nBest k: {best_k} with Silhouette Score = {best_score:.4f}")
    return best_k
```

```python
def cross_validate_clustering(X, model, n_splits=5, test_size=0.3):
    sil_scores, db_scores = [], []
    cv = ShuffleSplit(n_splits=n_splits, test_size=test_size, random_state=42)
    
    for i, (train_idx, test_idx) in enumerate(cv.split(X)):
        X_train, X_test = X[train_idx], X[test_idx]
        model.fit(X_train)
        
        # Predict on test set
        if hasattr(model, 'predict'):
            labels = model.predict(X_test)
        else:
            labels = model.fit_predict(X_test)
        
        sil = silhouette_score(X_test, labels)
        db = davies_bouldin_score(X_test, labels)
        
        sil_scores.append(sil)
        db_scores.append(db)
        print(f"Fold {i+1}: Silhouette = {sil:.4f}, DB Index = {db:.4f}")
    
    print("\nCross-Validation Results:")
    print(f"Mean Silhouette: {np.mean(sil_scores):.4f} ± {np.std(sil_scores):.4f}")
    print(f"Mean DB Index:   {np.mean(db_scores):.4f} ± {np.std(db_scores):.4f}")
    
    return sil_scores, db_scores
```

```python
# KMeans
kmeans_cv = KMeans(n_clusters=best_k2, random_state=42, n_init=20)
sil_scores_km, db_scores_km = cross_validate_clustering(X_scaled, kmeans_cv)

# Agglomerative
agg_cv = AgglomerativeClustering(n_clusters=best_k2, linkage='ward')
sil_scores_agg, db_scores_agg = cross_validate_clustering(X_scaled, agg_cv)
```
